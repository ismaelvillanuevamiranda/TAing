```python
def evaluate_model(name, model, X_train, X_test, y_train, y_test):
    """Train a model and show evaluation results with confusion matrix."""
    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred  = model.predict(X_test)

    if hasattr(model, 'predict_proba'):
        test_prob = model.predict_proba(X_test)[:, 1]
    else:
        d = model.decision_function(X_test)
        test_prob = (d - d.min()) / (d.max() - d.min())

    results = {
        'Model': name,
        'Train Accuracy': accuracy_score(y_train, train_pred),
        'Test Accuracy':  accuracy_score(y_test, test_pred),
        'Test Precision':  precision_score(y_test, test_pred),
        'Test Recall':     recall_score(y_test, test_pred),
        'Test ROC-AUC':    roc_auc_score(y_test, test_prob)
    }

    # Confusion matrix
    cm = confusion_matrix(y_test, test_pred)
    fig, ax = plt.subplots(figsize=(4, 3.5))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Benign', 'Malignant'])
    disp.plot(ax=ax, cmap='Blues')
    ax.set_title(f'Confusion Matrix: {name}', fontsize=12)
    plt.tight_layout()
    plt.show()

    # Print human-readable summary
    tn, fp, fn, tp = cm.ravel()
    print(f"  Correctly identified benign:    {tn}")
    print(f"  Correctly identified malignant: {tp}")
    print(f"  ❌ False alarms (benign called malignant): {fp}")
    print(f"  ❌ Missed cancers (malignant called benign): {fn}")

    return results

print("✅ Evaluation function ready.")
```

```python
# ▶️ RUN — Clean evaluation vs leaky evaluation by overlapping train and test sets
from sklearn.base import clone

# 1) Start from a proper split
X_clean_train, X_clean_test, y_clean_train, y_clean_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# 2) Train a baseline model on the clean training set
base_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=2000, random_state=42))
])

clean_model = clone(base_lr)
clean_model.fit(X_clean_train, y_clean_train)

clean_train_pred = clean_model.predict(X_clean_train)
clean_test_pred = clean_model.predict(X_clean_test)

clean_results = {
    'Scenario': 'Clean split',
    'Train Accuracy': accuracy_score(y_clean_train, clean_train_pred),
    'Test Accuracy': accuracy_score(y_clean_test, clean_test_pred),
    'ROC-AUC': roc_auc_score(y_clean_test, clean_model.predict_proba(X_clean_test)[:, 1])
}

# 3) Create a WRONG evaluation set by copying some training patients into the test set
rng = np.random.default_rng(42)
leak_fraction = 0.70   # 70% of the test set will be replaced by training patients
n_leak = int(len(X_clean_test) * leak_fraction)

train_idx_to_copy = rng.choice(X_clean_train.index.to_numpy(), size=n_leak, replace=False)
test_idx_to_keep = rng.choice(X_clean_test.index.to_numpy(), size=len(X_clean_test) - n_leak, replace=False)

X_leaky_test = pd.concat([
    X_clean_test.loc[test_idx_to_keep],
    X_clean_train.loc[train_idx_to_copy]
], axis=0)

y_leaky_test = pd.concat([
    y_clean_test.loc[test_idx_to_keep],
    y_clean_train.loc[train_idx_to_copy]
], axis=0)

# Shuffle so the copied rows are not obvious
shuffle_order = rng.permutation(len(X_leaky_test))
X_leaky_test = X_leaky_test.iloc[shuffle_order].reset_index(drop=True)
y_leaky_test = y_leaky_test.iloc[shuffle_order].reset_index(drop=True)

leaky_model = clone(base_lr)
leaky_model.fit(X_clean_train, y_clean_train)
leaky_test_pred = leaky_model.predict(X_leaky_test)

leaky_results = {
    'Scenario': f'Leaky test set ({leak_fraction:.0%} copied from training)',
    'Train Accuracy': accuracy_score(y_clean_train, leaky_model.predict(X_clean_train)),
    'Test Accuracy': accuracy_score(y_leaky_test, leaky_test_pred),
    'ROC-AUC': roc_auc_score(y_leaky_test, leaky_model.predict_proba(X_leaky_test)[:, 1])
}

comparison = pd.DataFrame([clean_results, leaky_results]).round(3)
display(comparison)

print(f"Number of leaked patients copied from training into test: {n_leak}")
print(f"Original test size: {len(X_clean_test)}")
print()
print("Why this is leakage:")
print("- In the clean split, test patients are completely unseen.")
print("- In the leaky split, part of the test set contains patients from the training set.")
print("- The model looks better because it is evaluated on examples it has already learned.")
```

### Key takeaway

```python
# ▶️ RUN
from sklearn.dummy import DummyClassifier
from sklearn.datasets import make_classification

# Create a dataset with 95% healthy, 5% diseased
X_imb, y_imb = make_classification(
    n_samples=2000, n_features=20, n_informative=8,
    n_redundant=4, weights=[0.95, 0.05], random_state=42
)

Xi_train, Xi_test, yi_train, yi_test = train_test_split(
    X_imb, y_imb, test_size=0.25, stratify=y_imb, random_state=42
)

# A "model" that ALWAYS predicts "healthy" — ignores all data
dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(Xi_train, yi_train)
dummy_pred = dummy.predict(Xi_test)

print('=== "Always predict healthy" dummy model ===')
print(f"  Accuracy:  {accuracy_score(yi_test, dummy_pred):.1%}")
print(f"  Precision: {precision_score(yi_test, dummy_pred, zero_division=0):.1%}")
print(f"  Recall:    {recall_score(yi_test, dummy_pred):.1%}")
print()
print("  ➡️  This model catches 0% of diseased patients!")
print(f"  ➡️  Yet it's {accuracy_score(yi_test, dummy_pred):.1%} accurate. Accuracy is LYING to us.")
```
